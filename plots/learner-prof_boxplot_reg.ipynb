this produces barplots and shows distribution of item frequency across texts in 5 (learner-prof) or 6 (pop-sci_mode) corpora. 
Normalized frqequencies are set out in a 'long' table under the variable name frequency; 
the other variables (or levels of this one) are competence (source,translation, reference) / mode (lectures, books) and language (EN/RU)

```python
import sys,codecs,os
import pandas as pd
import numpy as np
from numpy import median
import matplotlib.pyplot as plt
import seaborn as sns
```

each plot requires its own dataframe for best results!
competence(learner, professional); all frequences are normalize per 100 sentences

```python
df_conn = pd.read_csv("/home/masha/birmingham/2plot/tmx2boxplot_conn_long_learner-prof-ref.csv")
df_em = pd.read_csv("/home/masha/birmingham/2plot/tmx2boxplot_em_long_learner-prof-ref.csv")
```

```python
sns.set_style("whitegrid")
sns.set_context('paper')
```

```python
#colors = dict(source="lightgray", translation="darkgray", reference="k--")
ax = sns.boxplot(x="competence", y="frequency", hue="language", data=df_conn, palette="PRGn", showmeans=True) 
# for print version substitute palette="PRGn" for color="gray" change the outfile name!
plt.ylabel('cumulative frequency of connectives')

#plt.savefig('/home/masha/BOX_pop-ted.eps', format='eps')
plt.show()
```

```python
#colors = dict(source="lightgray", translation="darkgray", reference="k--")
ax = sns.boxplot(x="competence", y="frequency", hue="language", data=df_em, palette="PRGn", showmeans=True) 
# for print version substitute palette="PRGn" for color="gray" change the outfile name!
plt.ylabel('cumulative frequency of epistemic markers')

plt.savefig('/home/masha/emBOX_learner-prof.eps', format='eps')
plt.show()
```

the commands below plot regression for sources and targets at the two levels of competence 

they require other input files in the wide-table format

```python
df_conn = pd.read_csv("/home/masha/birmingham/2plot/tmx2regplot_conn_wide_learner-prof.csv", sep = ',')
df_em = pd.read_csv("/home/masha/birmingham/2plot/tmx2regplot_em_wide_learner-prof.csv", sep = ',')
sns.set_style("whitegrid")
sns.set_context('paper')
```

```python
g = sns.FacetGrid(data=df_conn, col='competence', size = 5, sharey=True)
#g.map(sns.kdeplot, 'source','target')
g.map(sns.regplot, 'source','target')

g.set(xlim=(0, 40))
g.set(ylim=(0, 30))

plt.tight_layout()

#plt.savefig('/home/masha/bivariateKDE_pop-ted.eps', format='eps')
plt.show()
```

```python
g = sns.FacetGrid(data=df_em, col='competence', size = 5, sharey=True)
#g.map(sns.kdeplot, 'source','target')
g.map(sns.regplot, 'source','target')

g.set(xlim=(0, 40))
g.set(ylim=(0, 30))

plt.tight_layout()

plt.savefig('/home/masha/emREG_learner-prof.eps', format='eps')
plt.show()
```
