```python
import pandas as pd
import numpy as np
from numpy import median
import matplotlib.pyplot as plt
import seaborn as sns
```

This script can only be used for pop-sci research where there is a reference corpus for each  pair of test corpora!

The underlying csv has an arbitrary num of rows for texts, a column with competence data and three columns for sources, targets and reference  
This produces two facets with histograms for sources/targets (represented in a wide-table style) for learners/professionals (competences define facets). The same hist for the reference is reproduced in each facet. The picture shows how often (y) each frequency interval (x) occurs in the dataset

the commands below are for pop-sci data only!

```python
df_conn = pd.read_csv("/home/masha/valencia/2plot/tmx2hist_conn_wide_pop-sci.csv", sep = ',')
df_em = pd.read_csv("/home/masha/valencia/2plot/tmx2hist_em_wide_pop-sci.csv", sep = ',')
sns.set_style("whitegrid")
sns.set_context('paper')
```

```python
g = sns.FacetGrid(data=df_conn, col='mode', size = 5)
'''
this produces difference slices for RNC as reference for each corpus!
replace .distplot with .kdeplot to buld a hist of probailities rather than actual frequencies
'''
g = g.map(sns.distplot, 'source', kde=False, bins=10, rug=False, hist_kws={"color": "lightgray","histtype": "step", "linewidth": 2,  "alpha": 1}) 
g = g.map(sns.distplot, 'target', kde=False, bins=10, rug=False, hist_kws={"color": "black","histtype": "step", "linewidth": 2,  "alpha": 1})
g = g.map(sns.distplot, 'reference', kde=False, bins=10, rug=False, hist_kws={"color": "red", "linestyle": "dashed","histtype": "step", "linewidth": 2,  "alpha": 1})

g.axes[0,0].set_xlabel('frequency of connectives')
g.axes[0,1].set_xlabel('frequency of connectives')
g.axes[0,0].set_ylabel('how often the frequency occurs in the data')
g.set(xlim=(0, 50))
g.add_legend() #gives the legend in the middle to the right. 
#без этой строки легенда графика не появляется
plt.legend()
plt.tight_layout()

#plt.savefig('/home/masha/hist_CONN_pop-sci.eps', format='eps')
plt.show()
```

```python
g = sns.FacetGrid(data=df_em, col='mode', size = 5)
'''
this produces difference slices for RNC as reference for each corpus!
replace .distplot with .kdeplot to buld a hist of probailities rather than actual frequencies
'''
g = g.map(sns.distplot, 'source', kde=False, bins=10, rug=False, hist_kws={"color": "lightgray","histtype": "step", "linewidth": 2,  "alpha": 1}) 
g = g.map(sns.distplot, 'target', kde=False, bins=10, rug=False, hist_kws={"color": "black","histtype": "step", "linewidth": 2,  "alpha": 1})
g = g.map(sns.distplot, 'reference', kde=False, bins=10, rug=False, hist_kws={"color": "red", "linestyle": "dashed","histtype": "step", "linewidth": 2,  "alpha": 1})

g.axes[0,0].set_xlabel('frequency of epistemic markers')
g.axes[0,1].set_xlabel('frequency of epistemic markers')
g.axes[0,0].set_ylabel('how often the frequency occurs in the data')
g.set(xlim=(0, 25))
#g1.add_legend() #gives the legend in the middle to the right. 
#без этой строки легенда графика не появляется
plt.legend()
plt.tight_layout()

plt.savefig('/home/masha/hist_EM_pop-sci.eps', format='eps')
plt.show()
```
